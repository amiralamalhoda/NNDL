```python
import numpy as np
import matplotlib.pyplot as plt

class Adaline:
    def __init__(self, learning_rate=0.1, init_weights = [0,0], init_bias = 0):
        self.learning_rate = learning_rate
        self.weights = init_weights
        self.bias = init_bias
    
    def train(self, x1, y1, x2, y2):
        # Create labeled data points
        positive_samples = [(a, b, 1) for a, b in zip(x1, y1)]
        negative_samples = [(a, b, -1) for a, b in zip(x2, y2)]
        data_points = positive_samples + negative_samples

        # Initialize variables for Adaline training
        epochs = 0
        iteration_count = 0
        converged = False

        while not converged:
            converged = True  # Assume convergence unless an update occurs
            epochs += 1

            for data in data_points:
                x, y, target = data
                z = self.predict((x,y))
                prediction = np.sign(z)
                error = target - prediction

                if error != 0:
                    converged = False  # Model hasn't converged yet
                    self.weights[0] += self.learning_rate * x * error
                    self.weights[1] += self.learning_rate * y * error
                    self.bias += self.learning_rate * error
                    iteration_count += 1

        print(f"The algorithm took {epochs} epochs to find the parameters and tried {iteration_count} different sets of parameters.")

    def predict(self, x):
        z = self.weights[0] * x[0] + self.weights[1] * x[1] + self.bias
        return np.sign(z)

    def plot_decision_boundary(self, xaxis=(-3, 3)):
        x = np.linspace(xaxis[0], xaxis[1], 100)
        y = (-self.weights[0] / self.weights[1]) * x - (self.bias / self.weights[1])
        plt.plot(x, y)

    def visualize_training(self, x1, y1, x2, y2, xaxis=(-3, 3)):
        plt.scatter(x1, y1, color='red')
        plt.scatter(x2, y2, color='blue')
        self.plot_decision_boundary(xaxis=xaxis)
        plt.xlim(xaxis[0], xaxis[1])
        plt.ylim(xaxis[0], xaxis[1])
        plt.xlabel('x1')
        plt.ylabel('x2')
        plt.grid()
        plt.show()
```
